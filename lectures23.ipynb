```python
class Portfolio:
    def __init__(self):
        self.instruments = []  # Crée une liste vide pour stocker les instruments financiers.

    def add_instrument(self, name, price):
        instrument = {'name': name, 'price': price}  # Crée un dictionnaire pour représenter l'instrument.
        self.instruments.append(instrument)  # Ajoute l'instrument à la liste du portefeuille.

    def total_value(self):
        total = 0
        for instrument in self.instruments:
            total += instrument['price']  # Accumule les prix de tous les instruments dans le portefeuille.
        return total

# Exemple d'utilisation :
my_portfolio = Portfolio()

# Ajout d'instruments au portefeuille
my_portfolio.add_instrument('Stock A', 100)
my_portfolio.add_instrument('Stock B', 150)
my_portfolio.add_instrument('Bond X', 500)

# Calcul de la valeur totale du portefeuille
portfolio_value = my_portfolio.total_value()
print(f"La valeur totale du portefeuille est : ${portfolio_value}")
```

```python
class CurrencyConverter:
    def __init__(self):
        # Initialisez un dictionnaire pour stocker les taux de conversion.
        # Les taux de conversion sont définis en fonction de la devise source vers la devise cible.
        self.rates = {}

    def add_rate(self, source_currency, target_currency, rate):
        # Ajoutez un taux de conversion au dictionnaire.
        self.rates[(source_currency, target_currency)] = rate

    def convert(self, amount, source_currency, target_currency):
        if (source_currency, target_currency) in self.rates:
            rate = self.rates[(source_currency, target_currency)]
            converted_amount = amount * rate
            return converted_amount
        else:
            return None  # Retournez None si le taux de conversion n'est pas disponible.

# Exemple d'utilisation :
currency_converter = CurrencyConverter()

# Ajout de taux de conversion
currency_converter.add_rate("USD", "EUR", 0.85)
currency_converter.add_rate("EUR", "JPY", 120.0)
currency_converter.add_rate("GBP", "USD", 1.35)

# Conversion d'un montant de USD en EUR
amount_usd = 100
source_currency = "USD"
target_currency = "EUR"
converted_amount = currency_converter.convert(amount_usd, source_currency, target_currency)

if converted_amount is not None:
    print(f"{amount_usd} {source_currency} équivaut à {converted_amount} {target_currency}")
else:
    print("Le taux de conversion n'est pas disponible.")
```

```python
import numpy as np

# Réglez la graine aléatoire pour la reproductibilité
np.random.seed(0)

# Générez des rendements quotidiens aléatoires (échantillon de 1000 nombres)
daily_returns = np.random.normal(0.0001, 0.02, 1000)

# Initialisez la liste des prix de l'action avec une valeur de départ
stock_prices = [100]

# Calculez les prix de l'action sur la base des rendements quotidiens
for r in daily_returns:
    new_price = stock_prices[-1] * (1 + r)
    stock_prices.append(new_price)

# Imprimez le dernier prix de l'action
print(stock_prices[-1])
```
